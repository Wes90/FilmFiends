```python
import pandas as pd
```

```python
file = 'Resources/movies_metadata.csv'
```

```python
df = pd.read_csv(file)
df.head()
```

```python
df.count()
```

```python
df = df[["title", "budget", "revenue", "vote_average"]]
```

```python
df.head()
```

```python
df.count()
```

```python
df = df.dropna(how='any')
df.head()
```

```python
df.count()
```

```python
df = df.rename(columns={"title":"Title", "budget":"Budget", "revenue":"Revenue","vote_average":"Average Vote"})
```

```python
df.head()
```

```python
indexbudget = df[df['Budget'] == '0'].index
df.drop(indexbudget , inplace=True)
```

```python
indexrevenue = df[df['Revenue'] == 0.0].index
df.drop(indexrevenue , inplace=True)
```

```python
df.head()
```

```python
df.count()
```

```python
df.info()
```

```python
df["Budget"] = df["Budget"].astype(float)
```

```python
gross_revenue = df["Revenue"] + df["Budget"]
```

```python
df["Gross Revenue"] = gross_revenue
```

```python
df.head()
```

```python
df = df[["Title", "Budget","Revenue","Gross Revenue", "Average Vote"]]
df.head() 
```
