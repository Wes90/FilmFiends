```python
import pandas as pd
```

```python
file = 'Resources/movies_metadata.csv'
```

```python
df = pd.read_csv(file)
df.head()
```

```python
df.count()
```

```python
df = df[["title","release_date" ,"budget", "revenue", "vote_average"]]
```

```python
df.head()
```

```python
df.count()
```

```python
df = df.dropna(how='any')
df.head()
```

```python
df.count()
```

```python
df = df.rename(columns={"title":"Title","release_date":"Year" , "budget":"Budget", "revenue":"Revenue","vote_average":"Average Vote"})
```

```python
df.head()
```

```python
indexbudget = df[df['Budget'] == '0'].index
df.drop(indexbudget , inplace=True)
```

```python
indexrevenue = df[df['Revenue'] == 0.0].index
df.drop(indexrevenue , inplace=True)
```

```python
df.head()
```

```python
df.count()
```

```python
df.info()
```

```python
df["Budget"] = df["Budget"].astype(float)
```

```python
gross_revenue = df["Revenue"] + df["Budget"]
```

```python
df["Gross Revenue"] = gross_revenue
```

```python
df.head()
```

```python
df = df[["Title","Year" ,"Budget","Revenue","Gross Revenue", "Average Vote"]]
df.head() 
```

```python
new = df["Year"].str.split("-", n=1, expand = True)
df["Year"] = new[0]
```

```python
df.head()
```

```python
df.info()
```

```python
df["Year"] = df["Year"].astype(int)
```

```python
df = df[(df["Year"]>1989)]
```

```python
df = df[(df["Year"]<2000)]
```

```python
df.head(10)
```

```python
df.to_csv('meta_data_cleaned_final.csv')
```
